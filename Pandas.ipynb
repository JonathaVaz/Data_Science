# ***Conceitos Estatísticos ( IMPORTANTE! )***

```python
# Usando de Exemplo Dados Fictísio em que X corresponde a uma série de números, e atríbuimos a um dataframe pandas

dados = { 
         'X' : [1,2,3,10,0,19] #,4,5,6,7,8,9,10,11
}

dados = pandas.DataFrame(dados)
```

# ***Fazendo Filtros na Base***

```python
# Visualizando a Base novamente
base.head()
```

```python
# Verificando Clientes que Pagaram mais de 500 Libras
base[base.Fare > 500]
```

```python
# Verificando se tiveram clientes que pagaram menos de 5 libras
base[base.Fare < 5]
```

### **Podemos Fazer Filtros usando E / OU**

```python
# Verificando se Alguém viajou com esposa e/ou filhos
# E = & 
# OU = |

base[(base.Parch > 1) | (base.SibSp > 1)].head(10)
```

```python
# Verificando se Alguém viajou com esposa e/ou filhos
# E = & 
# OU = |

base[(base.Parch > 1) & (base.SibSp > 1)].head(10)
```

# Uma Forma Útil de Fazer Seleção de Dados é usando o .loc() ou o .iloc()

```python
# O .loc() permite fazer a busca da mesma forma acima
base.loc[(base.Parch > 1) & (base.SibSp > 1)].head(5)
```

```python
# Também permite usar argumentos lógicos
base.loc[(base.Parch > 1) | (base.SibSp > 1)].head(5)
```

```python
# Também Permite filtrar colunas de forma prática
base.loc[(base.Parch > 1) | (base.SibSp > 1),['PassengerId','Parch','SibSp']].head(5)
```

# Já o .iloc() vai usar o índice para filtrar

```python
# Buscando as linhas de 30 (inclusive) a 40 (exclusive)
base.iloc[30:40]
```

```python
# Também posso usar para buscar apenas colunas específicas
base.iloc[30:40, 3:6]
```
