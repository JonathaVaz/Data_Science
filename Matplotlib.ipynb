#### Vamos Criar uma Série de Gráficos com Diferentes Análises Visuais, sendo de Musicas, Jogadores da NBA, Carros e Preços de Imoveis 

```python
# Importando o Matplotlib
import matplotlib.pyplot as plt
```

# ***Gêneros Musicais***

```python
# Criando Variáveis Correspondentes a de Gêneros Musicais e Quantidade de Musicas Baixadas
genero = ["Rock", "Hip Hop", "Pop", "Jazz", "Reggae"]
quantidade_msc = [330, 570, 133, 190, 260]
```

```python
# Definindo a Lista de Cores de cada Gênero
cores = plt.color=('firebrick','mediumpurple','springgreen','palegreen','royalblue')
```

```python
# Definindo o Fundo do Gráfico - Todo Preto
plt.style.use('dark_background')
```

```python
# Criando o Gráfico

plt.title("Quantidade dos Gêneros Musicais baixados")
plt.bar(genero, quantidade_msc, color=cores, edgecolor='white')
```

```python
# Melhorando a Visualização de Dados

plt.figure(figsize=(13, 8))

plt.title("Quantidade dos Gêneros Musicais baixados\n", fontsize=17, fontweight='bold', style='italic')
plt.bar(genero, quantidade_msc, color=cores, edgecolor='white')
plt.xticks(fontsize=12, fontweight='bold', style='italic')
plt.yticks(fontsize=12, fontweight='bold', style='italic')
plt.xlabel('\nGêneros Musicais', fontsize=15, fontweight='bold', style='italic')
plt.ylabel('Quantidade de Músicas\n', fontsize=16, fontweight='bold', style='italic')
plt.show()
```

```python
# Colocando os Marcadores de Contagens em cima das Barras 

plt.figure(figsize=(13, 8))

plt.title("Quantidade dos Gêneros Musicais baixados\n", fontsize=17, fontweight='bold', style='italic')
plt.bar(genero, quantidade_msc, color=cores, edgecolor='white')
plt.xticks(fontsize=12, fontweight='bold', style='italic')
plt.yticks(fontsize=12, fontweight='bold', style='italic')
plt.xlabel('\nGêneros Musicais', fontsize=15, fontweight='bold', style='italic')
plt.ylabel('Quantidade de Músicas\n', fontsize=16, fontweight='bold', style='italic')
for x,y in zip(genero,quantidade_msc):
    plt.text(x, y+0.05, '%d' % y, ha='center' , va= 'bottom', fontsize=12, fontweight='bold', style='italic')
plt.show()
```

```python
# Criando Gráfico de "Pizza" em Porcentagem para Saber o Percentual de cada Musica Baixada

plt.figure(figsize=(10,7))

plt.title('Quantidade dos Gêneros Musicais baixados\n', fontsize=15, fontweight='bold', style='italic')
plt.pie(quantidade_msc, labels=genero, autopct='%1.1f%%', colors=(cores))
plt.ylabel('')
plt.xlabel('Gêneros Musicais', fontsize=13, fontweight='bold', style='italic')
plt.show()
```

# ***Jogadores da NBA***

```python
# Criando Variáveis Correspondentes a de Jogadores da NBA e Quantidade de Pontos Feitos no Último Jogo
jogador_nba = ["Kevin Durant", "Antony Davis", "Lebron James", "Bam Adebayo", "Shai Gilgeous-Alexander"]
quantidade_pontos = [32, 37, 44, 23, 35]
```

```python
# Definindo a Lista de Cores de cada Jogador
cores_nba = plt.color=('darkorange','mediumturquoise','slateblue','palevioletred','lightgreen')
```

```python
# Definindo o Fundo do Gráfico - Todo Preto
plt.style.use('dark_background')
```

```python
# Criando o Gráfico

plt.figure(figsize=(12, 7))

plt.title("Jogadores da NBA e a Quantidade de Pontos feitos no Último Jogo\n", fontsize=15, fontweight='bold', style='italic')
plt.bar(jogador_nba, quantidade_pontos, color=cores_nba, edgecolor='white')
plt.xticks(fontsize=10, fontweight='bold', style='italic')
plt.yticks(fontsize=10, fontweight='bold', style='italic')
plt.xlabel('\nJogadores', fontsize=12, fontweight='bold', style='italic')
plt.ylabel('Quantidade de Pontos\n', fontsize=12, fontweight='bold', style='italic')
for x,y in zip(jogador_nba,quantidade_pontos):
    plt.text(x, y+0.05, '%d' % y, ha='center' , va= 'bottom', fontsize=12, fontweight='bold', style='italic')
plt.show()
```

```python
# Criando Gráfico de "Pizza" em Porcentagem para Saber o Percentual de cada Jogador

plt.figure(figsize=(10,7))

plt.title('Jogadores da NBA e a Quantidade de Pontos feitos no Último Jogo\n', fontsize=15, fontweight='bold', style='italic')
plt.pie(quantidade_pontos, labels=jogador_nba, autopct='%1.1f%%', colors=(cores_nba))
plt.ylabel('')
plt.xlabel('Jogadores', fontsize=13, fontweight='bold', style='italic')
plt.show()
```

# ***Carros e Velocidade Final***

### - Usando o Seaborn para colocar Paleta de Cores
### - O Seaborn também é uma Biblioteca de Visualização assim como o Matplotlib, porém, mais usada para Visualização de Cores e Métodos Estatíscos para Visualização
### - Iremos usar apenas para Paleta de Cores

```python
# Importando o Seaborn
import seaborn as sns
```

```python
# Criando Variáveis Correspondentes a Carros e Velocidade Final de cada um
carros = ["Chevelle SS", "Datsun 240", "Lancer Evolution VII", "Corvette z06", "Viper SRT10"]
velocidade_final = [278, 295, 330, 350, 386]
```

```python
# Definindo a Paleta de Cores
cores_carros = sns.color_palette('gnuplot2')
```

```python
# Definindo o Fundo do Gráfico - Todo Preto
plt.style.use('dark_background')
```

```python
# Criando o Gráfico

plt.figure(figsize=(12, 7))

plt.title("Velocidade Final dos Veículos Registrados\n", fontsize=15, fontweight='bold', style='italic')
plt.bar(carros, velocidade_final, color=cores_carros, edgecolor='white')
plt.xticks(fontsize=10, fontweight='bold', style='italic')
plt.yticks(fontsize=10, fontweight='bold', style='italic')
plt.xlabel('\nCarros', fontsize=12, fontweight='bold', style='italic')
plt.ylabel('Velocidade Final\n', fontsize=12, fontweight='bold', style='italic')
for x,y in zip(carros,velocidade_final):
    plt.text(x, y+0.05, '%d' % y, ha='center' , va= 'bottom', fontsize=12, fontweight='bold', style='italic')
plt.show()
```

```python
# Criando Gráfico de "Pizza" em Porcentagem para Saber o Percentual da Velocidade Final de cada Veículo

plt.figure(figsize=(10,7))

plt.title('Velocidade Final dos Veículos Registrados\n', fontsize=15, fontweight='bold', style='italic')
plt.pie(velocidade_final, labels=carros, autopct='%1.1f%%', colors=(cores_carros))
plt.ylabel('')
plt.xlabel('Carros', fontsize=13, fontweight='bold', style='italic')
plt.show()
```

# ***Preços de Imovéis de Acordo com o Tipo***

```python
# Criando Variáveis Correspondentes a Imovéis e Preço de Acordo com a Localidade
imoveis = ["Triplex", "Apartamento", "Casa de Praia", "Casa", "Mansão"]
preco = [350.000, 220.000, 520.000, 370.000, 890.000]
```

```python
# Definindo o Fundo do Gráfico - Todo Preto
plt.style.use('dark_background')
```

```python
# Criando o Gráfico

plt.figure(figsize=(12, 7))

plt.title("Preços de Imovéis de Acordo com o Tipo do Imovél\n", fontsize=15, fontweight='bold', style='italic')
plt.bar(imoveis, preco, color=cores_imoveis, edgecolor='white')
plt.xticks(fontsize=10, fontweight='bold', style='italic')
plt.yticks(fontsize=10, fontweight='bold', style='italic')
plt.xlabel('\nImovéis', fontsize=12, fontweight='bold', style='italic')
plt.ylabel('Preços\n', fontsize=12, fontweight='bold', style='italic')
for x,y in zip(imoveis,preco):
    plt.text(x, y+0.05, '%d' % y, ha='center' , va= 'bottom', fontsize=12, fontweight='bold', style='italic')
plt.show()
```

```python
# Criando Gráfico de "Pizza" em Porcentagem para Saber o Percentual dos Valores dos Imovéis

plt.figure(figsize=(10,7))

plt.title('Preços de Imovéis de Acordo com o Tipo do Imovél\n', fontsize=15, fontweight='bold', style='italic')
plt.pie(preco, labels=imoveis, autopct='%1.1f%%', colors=(cores_imoveis))
plt.ylabel('')
plt.xlabel('Imovéis', fontsize=13, fontweight='bold', style='italic')
plt.show()
```
